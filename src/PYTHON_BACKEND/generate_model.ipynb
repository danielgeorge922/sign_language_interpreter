```python
import pandas as pd
import json

# Paths to the JSON files
main_path = 'C:\\Users\\danie\\sign-language-data\\'
wlasl_json_path = main_path + 'WLASL_v0.3.json'
nslt_100_path = main_path + 'nslt_100.json'
nslt_300_path = main_path + 'nslt_300.json'
nslt_1000_path = main_path + 'nslt_1000.json'
nslt_2000_path = main_path + 'nslt_2000.json'
video_path = main_path + 'videos\\'



# Load JSON files
with open(wlasl_json_path, 'r') as file:
    wlasl_data = json.load(file)
    print(f"Loaded WLASL data: {len(wlasl_data)} entries")
    
with open(nslt_100_path, 'r') as file:
    nslt_100_data = json.load(file)
    print(f"Loaded NSLT 100 data: {len(nslt_100_data)} entries")

with open(nslt_300_path, 'r') as file:
    nslt_300_data = json.load(file)
    print(f"Loaded NSLT 300 data: {len(nslt_300_data)} entries")
    
with open(nslt_1000_path, 'r') as file:
    nslt_1000_data = json.load(file)
    print(f"Loaded NSLT 1000 data: {len(nslt_1000_data)} entries")
    
with open(nslt_2000_path, 'r') as file:
    nslt_2000_data = json.load(file)
    print(f"Loaded NSLT 2000 data: {len(nslt_2000_data)} entries")



# Combine the NSLT data into a single dictionary
nslt_data = {**nslt_100_data, **nslt_300_data, **nslt_1000_data, **nslt_2000_data}
print(f"Combined NSLT data: {len(nslt_data)} entries")

# Initialize lists for the DataFrame
video_ids, subsets, actions, glosses, sources, signer_ids, image_paths, bboxes, fps_list, frame_starts, frame_ends, instance_ids, variation_ids = [], [], [], [], [], [], [], [], [], [], [], [], []



# Process WLASL data
for gloss_data in wlasl_data:
    gloss = gloss_data['gloss']
    for instance in gloss_data['instances']:
        video_id = instance['video_id']
        subset = instance['split']
        source = instance['source']
        signer_id = instance['signer_id']
        action = nslt_data.get(video_id, {}).get('action', None)
        image_path = video_path + video_id + ".mp4"
        bbox = instance.get('bbox', None)
        fps = instance.get('fps', None)
        frame_start = instance.get('frame_start', None)
        frame_end = instance.get('frame_end', None)
        instance_id = instance.get('instance_id', None)
        variation_id = instance.get('variation_id', None)

        video_ids.append(video_id)
        subsets.append(subset)
        actions.append(action)
        glosses.append(gloss)
        sources.append(source)
        signer_ids.append(signer_id)
        image_paths.append(image_path)
        bboxes.append(bbox)
        fps_list.append(fps)
        frame_starts.append(frame_start)
        frame_ends.append(frame_end)
        instance_ids.append(instance_id)
        variation_ids.append(variation_id)



# Create the DataFrame
combined_df = pd.DataFrame({
    'video_id': video_ids,
    'subset': subsets,
    'action': actions,
    'gloss': glosses,
    'source': sources,
    'signer_id': signer_ids,
    'image_path': image_paths,
    'bbox': bboxes,
    'fps': fps_list,
    'frame_start': frame_starts,
    'frame_end': frame_ends,
    'instance_id': instance_ids,
    'variation_id': variation_ids
})

print("DataFrame created successfully!")
```

```python
import cv2
import numpy as np
import pandas as pd
import os
from tensorflow.keras.utils import to_categorical
import mediapipe as mp

# Initialize MediaPipe Holistic model
mp_holistic = mp.solutions.holistic

# Function to extract keypoints
def extract_keypoints(results):
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose, face, lh, rh])

# Function to process frames
def process_frames(frames):
    keypoints = []
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
        for frame in frames:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = holistic.process(frame_rgb)
            keypoints.append(extract_keypoints(results))
    return np.array(keypoints)

# Function to extract frames from a video
def extract_frames(video_path, num_frames=30):
    cap = cv2.VideoCapture(video_path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)
    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            frame = cv2.resize(frame, (224, 224))
            frames.append(frame)
        else:
            break
    cap.release()
    return np.array(frames)

# Create directory to store keypoints
keypoints_dir = r'C:\\Users\\danie\\sign-language-data\\keypoints'
os.makedirs(keypoints_dir, exist_ok=True)

# Add column for keypoints paths
combined_df['keypoints_npy_path'] = combined_df['video_id'].apply(lambda x: os.path.join(keypoints_dir, f"{x}.npy"))

# Function to create datasets in batches and save intermediate results
def create_datasets(df, batch_size=100):
    batch_counter = 0
    for start in range(0, len(df), batch_size):
        sequences, labels = [], []
        end = min(start + batch_size, len(df))
        batch_df = df.iloc[start:end]
        for _, row in batch_df.iterrows():
            video_path = row['image_path']
            frames = extract_frames(video_path, num_frames=30)
            if len(frames) == 30:  # Ensure all videos have the same number of frames
                keypoints = process_frames(frames)
                sequences.append(keypoints)
                labels.append(row['action'][0])  # Use the first action as label

        np.save(os.path.join(keypoints_dir, f'batch_sequences_{batch_counter}.npy'), np.array(sequences))
        np.save(os.path.join(keypoints_dir, f'batch_labels_{batch_counter}.npy'), np.array(labels))
        batch_counter += 1

# Splitting the DataFrame into train, val, and test
train_df = combined_df[combined_df['subset'] == 'train']
val_df = combined_df[combined_df['subset'] == 'val']
test_df = combined_df[combined_df['subset'] == 'test']

# Process and save datasets in batches
create_datasets(train_df)
create_datasets(val_df)
create_datasets(test_df)

print("Data processing complete. Batches saved as .npy files.")
```

```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical

# Function to load keypoints from npy file
def load_keypoints(npy_path):
    return np.load(npy_path)

# Prepare dataset by loading keypoints and labels
def prepare_dataset(df):
    X, y = [], []
    for idx, row in df.iterrows():
        keypoints = load_keypoints(row['keypoints_npy_path'])
        X.append(keypoints)
        y.append(row['action'][0])  # Assuming first action label is the one to predict
    return np.array(X), np.array(y)

# Add column for keypoints paths (in case it wasn't already added)
keypoints_dir = r'C:\\Users\\danie\\sign-language-data\\keypoints3'
combined_df['keypoints_npy_path'] = combined_df['video_id'].apply(lambda x: os.path.join(keypoints_dir, f"{x}.npy"))

# Filter DataFrame to only include rows with existing keypoints files
combined_df = combined_df[combined_df['keypoints_npy_path'].apply(lambda x: os.path.exists(x))]

# Splitting the DataFrame into train, val, and test sets
train_df = combined_df[combined_df['subset'] == 'train']
val_df = combined_df[combined_df['subset'] == 'val']
test_df = combined_df[combined_df['subset'] == 'test']

# Prepare datasets
X_train, y_train = prepare_dataset(train_df)
X_val, y_val = prepare_dataset(val_df)
X_test, y_test = prepare_dataset(test_df)

# One-hot encode labels
y_train = to_categorical(y_train, num_classes=2423)
y_val = to_categorical(y_val, num_classes=2423)
y_test = to_categorical(y_test, num_classes=2423)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import mediapipe as mp

# Initialize MediaPipe Holistic model
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils

# Load the trained model
model = load_model('action_model.h5')

# Load the class labels
class_labels = combined_df['gloss'].unique()
num_classes = len(class_labels)

# Function to extract keypoints from the holistic model results
def extract_keypoints(results):
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33 * 4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468 * 3)
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21 * 3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21 * 3)
    return np.concatenate([pose, face, lh, rh])

# Initialize video capture
cap = cv2.VideoCapture(0)

# Initialize an empty list to store keypoints from each frame
sequence = []

# Define the number of frames to be used for prediction
sequence_length = 30

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convert the frame to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Process the frame to get holistic results
        results = holistic.process(frame_rgb)

        # Draw the landmarks on the frame
        if results.pose_landmarks:
            mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
        if results.face_landmarks:
            mp_drawing.draw_landmarks(frame, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION)
        if results.left_hand_landmarks:
            mp_drawing.draw_landmarks(frame, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        if results.right_hand_landmarks:
            mp_drawing.draw_landmarks(frame, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

        # Extract keypoints from the results
        keypoints = extract_keypoints(results)
        sequence.append(keypoints)

        # If we have collected enough frames, make a prediction
        if len(sequence) == sequence_length:
            input_data = np.expand_dims(sequence, axis=0)
            prediction = model.predict(input_data)
            predicted_class = np.argmax(prediction, axis=1)[0]
            
            # Handle case where predicted_class is out of bounds
            if predicted_class < num_classes:
                predicted_sign = class_labels[predicted_class]
            else:
                predicted_sign = "Unknown"

            # Display the predicted sign on the frame
            cv2.putText(frame, f'Predicted Sign: {predicted_sign}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

            # Remove the oldest frame from the sequence
            sequence.pop(0)

        # Show the frame with the prediction
        cv2.imshow('Sign Language Detection', frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release the video capture and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import pandas as pd
import os
from tensorflow.keras.utils import to_categorical
import mediapipe as mp

# Initialize MediaPipe Holistic model
mp_holistic = mp.solutions.holistic

# Function to extract keypoints
def extract_keypoints(results):
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21 * 3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21 * 3)
    return np.concatenate([lh, rh])

# Function to process frames
def process_frames(frames):
    keypoints = []
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
        for frame in frames:
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = holistic.process(frame_rgb)
            keypoints.append(extract_keypoints(results))
    return np.array(keypoints)

# Function to extract frames from a video
def extract_frames(video_path, num_frames=30):
    cap = cv2.VideoCapture(video_path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)
    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            frame = cv2.resize(frame, (224, 224))
            frames.append(frame)
        else:
            break
    cap.release()
    return np.array(frames)

# Create directory to store keypoints
keypoints_dir = r'C:\\Users\\danie\\sign-language-data\\new_keypoints'
os.makedirs(keypoints_dir, exist_ok=True)

# Add column for keypoints paths
combined_df['keypoints_npy_path'] = combined_df['video_id'].apply(lambda x: os.path.join(keypoints_dir, f"{x}.npy"))

# Function to create datasets in batches and save intermediate results
def create_datasets(df, batch_size=100):
    batch_counter = 0
    for start in range(0, len(df), batch_size):
        sequences, labels = [], []
        end = min(start + batch_size, len(df))
        batch_df = df.iloc[start:end]
        for _, row in batch_df.iterrows():
            video_path = row['image_path']
            frames = extract_frames(video_path, num_frames=30)
            if len(frames) == 30:  # Ensure all videos have the same number of frames
                keypoints = process_frames(frames)
                sequences.append(keypoints)
                labels.append(row['action'][0])  # Use the first action as label

        np.save(os.path.join(keypoints_dir, f'batch_sequences_{batch_counter}.npy'), np.array(sequences))
        np.save(os.path.join(keypoints_dir, f'batch_labels_{batch_counter}.npy'), np.array(labels))
        print(f"Batch {batch_counter} saved.")
        batch_counter += 1

# Splitting the DataFrame into train, val, and test
train_df = combined_df[combined_df['subset'] == 'train']
val_df = combined_df[combined_df['subset'] == 'val']
test_df = combined_df[combined_df['subset'] == 'test']

# Process and save datasets in batches
print("Processing training data...")
create_datasets(train_df)
print("Processing validation data...")
create_datasets(val_df)
print("Processing test data...")
create_datasets(test_df)

print("Data processing complete. Batches saved as .npy files.")

# Load keypoints and labels
def load_dataset(batch_sequences_dir, batch_labels_dir):
    sequences = []
    labels = []
    for seq_file, label_file in zip(sorted(os.listdir(batch_sequences_dir)), sorted(os.listdir(batch_labels_dir))):
        if seq_file.endswith('.npy') and label_file.endswith('.npy'):
            sequences.append(np.load(os.path.join(batch_sequences_dir, seq_file)))
            labels.append(np.load(os.path.join(batch_labels_dir, label_file)))
    return np.concatenate(sequences, axis=0), np.concatenate(labels, axis=0)

# Load training, validation, and test datasets
print("Loading training data...")
X_train, y_train = load_dataset(os.path.join(keypoints_dir, 'batch_sequences_train'), os.path.join(keypoints_dir, 'batch_labels_train'))
print("Loading validation data...")
X_val, y_val = load_dataset(os.path.join(keypoints_dir, 'batch_sequences_val'), os.path.join(keypoints_dir, 'batch_labels_val'))
print("Loading test data...")
X_test, y_test = load_dataset(os.path.join(keypoints_dir, 'batch_sequences_test'), os.path.join(keypoints_dir, 'batch_labels_test'))

# One-hot encode labels
y_train = to_categorical(y_train, num_classes=2423)
y_val = to_categorical(y_val, num_classes=2423)
y_test = to_categorical(y_test, num_classes=2423)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"y_val shape: {y_val.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

print("Dataset preparation complete.")
```
